```python
import dep_checker
dep_checker.check({"plotly","pandas","numpy"})
import numpy as np
purchasePrice = 2.4e6
numberUnits   = 2
          # 2bd2ba  2bd1.5ba 2bd1.5ba +1bd/1ba
rent      = 2915  + 2600     +1800    +1750 # current
rentPF    = 3600  + 2800     +2900    +2000 # market
highRent  = 400   + 300      +300     +150
midRent   = 300   + 200      +200     +100
lowRent   = 200   + 150      +150     +100

highVac  = 80
highDays = 18*7 # 12 summer, 2 - christmas break,1- thanksgiving, 1-easter, 2-spring break
midVac   = 70
midDays  = 100
lowVac   = 70
lowDays  = 365-highDays-midDays

overhead = 20 # short term rental and marking fee, net of rent

ltv         = 80
loanRatePer = 3.5



"""
create two business 1st own real estate (annuity+equity) 2nd business runs a
   would need to have escrow account for deferred maintenance
first business create NNN lease 
second business makes profit runs hotel minus lease, fees, maintenance
"""
capSpread   = 1#unit [%] 
taxes       = purchasePrice * 1.25/100.
insurance   = purchasePrice * .2/100.
maintenance = numberUnits * 5000 + purchasePrice * 1/100.
lease        = purchasePrice *(loanRatePer + capSpread)/100+taxes+insurance
print("""
cap    : {CAP}
lease  : {LEASE}({LEASE_M})
taxes  : {TAX}
income : {EBITDA}
DebtSrv: {DEBTSRV}
""".format(
        CAP=loanRatePer + capSpread,
        LEASE=lease,
        LEASE_M = lease/12.,
        TAX=taxes , 
        EBITDA=lease-taxes-insurance,
        DEBTSRV = purchasePrice*ltv/100 * loanRatePer/100
))

totalRent =   highRent * highVac/ 100. * highDays + midRent * midVac/ 100. * midDays + lowRent * lowVac/ 100. * lowDays
totalRent *=  1#numberUnits
managementFee= 15# unit [%]
income    =   totalRent  -totalRent*managementFee/100- maintenance# only consider mainten


print("""
total rent : {TR}
maintenance: {MAIN}
magagement : {MGMT}
income     : {INCOME}
insurance  : {INSR}

""".format(
    TR=totalRent,
    MAIN=maintenance ,
    MGMT=totalRent*managementFee/100,
    INCOME=income-lease,
    INSR=insurance));
"""
    P/E        : {PE}
    Down Pay/E : {DPE}
    PE=purchasePrice/income,  
    DPE=(purchasePrice*ltv/100)/income))
    """
''
```
